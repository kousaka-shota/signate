# 集合
何らかの集まりのことを集合と呼ぶ。例えば、サイコロの出る目であれば1から6までの自然数となるので

$$
1,2,3,4,5,6
$$
である。各集合の要素のことを元と呼ぶ。集合として表記する時は

$$
\left\{1,2,3,4,5,6 \right\}
$$
のように表す。次にこの集合を$A$とし、Aに含まれる何らかの要素を$x$とするとき、

$$
A=\left\{x \left. \right| 1,2,3,4,5,6 \right\}
$$
のように書く。また、$x$は$A$の集合であるということを表すとき

$$
x \in A
$$
のように書く。よく使われる表現として$x$がどのような性質の数かを明記する際によく使われる。

$$
\begin{align}
x &\in \mathbb{N} & x\mbox{は自然数} \\
x &\in \mathbb{Z} & x\mbox{は整数} \\
x &\in \mathbb{Q} & x\mbox{は有理数} \\
x &\in \mathbb{R} & x\mbox{は実数} \\
x &\in \mathbb{C} & x\mbox{は複素数} \\
\end{align}
$$
反対に$x$が$A$の要素でないときは

$$
x \not \in A 
$$
と表す。また考えている対象に対して$A$に含まれない部分のことを

$$
\bar{A} \\
A^{c}
$$
と表す。高校数学では$\bar{A}$とかくが、上つきバーは平均などで用いられることが多いので、$A^{c}$を使うことが多い。
集合に何も元がないことを空集合と呼び

$$
\phi
$$

また、$B=\left\{1,3,5 \right\}$
としたときに、$B$の要素は全て$A$に含まれている。このとき、$B$は$A$の部分集合と呼び

$$
B \subset A \\
A \supset B
$$
のように表す。$B=\left\{1,2,3,4,5,6 \right\}$としたときは、$A$と$B$は等しいので

$$
B=A
$$
と表す。集合$B=\left\{1,7,8 \right\}$とする。集合$A$と$B$を合わせたものを和集合と呼び、

$$
A \cup B = \left\{1,2,3,4,5,6,7,8 \right\}
$$
と表す。集合$A$と$B$ともに含まれるものの集合を積集合と呼び

$$
A \cap B = \left\{1\right\}
$$
また、積集合が空集合(共通部分がない)二つの集合のことを互いに排反と呼ぶ。

$$
A \cap B = \phi
$$
また、集合の要素の数を$n\left( \cdot \right)$または$\left| \cdot \right|$と表す。

$$
A=\left\{x \left. \right| 1,2,3,4,5,6 \right\} \\
\left| A \right| = \left| \left\{x \left. \right| 1,2,3,4,5,6 \right\} \right|  =6
$$
また集合$A$,$B$,$C$について以下の関係がなりたつ。

$$
\begin{align}
A \cap \left( B \cup C \right) &= \left( A \cap B \right) \cup \left( A \cap C \right) \\
A \cup \left( B \cap C \right) &= \left( A \cup B \right) \cap \left( A \cup C \right) \\
\end{align}
$$

# 関数
ある数の集合$\left\{x | x_{1}, x_{2}, \cdots, x_{n} \right\}$に対して、以下のような対応があるとする。
$$
\begin{align}
x_{1} \rightarrow & y_{1} \\
x_{2} \rightarrow & y_{2} \\
\cdots \\
x_{n} \rightarrow & y_{n} \\
\end{align}
$$

すなわち入力の集合$\left\{x | x_{1}, x_{2}, \cdots, x_{n} \right\}$に対応する出力集合が$\left\{y| y_{1}, y_{2}, \cdots, y_{n} \right\}$
であるとき、$y$を$x$の関数であるといい$y=f \left( x \right)$と表される。このとき$x$を独立変数、$y$を従属変数という。

### 例
独立変数を自然数$\left\{1, 2, 3, \cdots \right\}$としたときに以下のような関係があるとする。

|    |    |    |    |         |
|----|----|----|----|---------|
|$x$ |$1$ |$2$ |$3$ |$\cdots$ |
|$y$ |$2$ |$4$ |$6$ |$\cdots$ |

このとき$y=f \left( x \right)=2x$と表す。

# 数列と和、積の表現

何らかの順序を持った数の並びを数列と呼び

$$
a_{1},a_{2}, \cdots , a_{n}
$$

のように表す。いまこの数列の和を次のように表す

$$
\underbrace{a_{1}+a_{2}+ \cdots + a_{n}}_{n\mbox{個}} = \sum_{i=1}^{n} a_{i}
$$
この右辺は1項から$n$項まで全て足すという意味である。

また次のような表現をすることもある。集合$A= \left\{a_{1}, a_{2}, \cdots , a_{n} \right\}$の要素とみなして

$$
\sum_{a_{i} \in A} a_{i} = a_{1}+a_{2}+ \cdots + a_{n}
$$

と書くようなこともある。足す範囲が明らかな場合は

$$
\sum a_{i}
$$
と単に記載する場合もある。
### 例
$$
\begin{align}
\sum_{k=3}^{10} a_{k} =& a_{3} + a_{4} + a_{5} + a_{6} + a_{7} + a_{8} + a_{9} + a_{10} \\
\sum_{k=2}^{n} a_{k} =& \underbrace{a_{2} + a_{3} + \cdots + a_{n}}_{n-1\mbox{個}} \\
\sum_{k=1}^{6}\left( k+2 \right) =& 3 + 4 + 5 + 6 + 7 + 8 \\
\sum_{i=1}^{4} a_{k}   =&  a_{k} + a_{k} + a_{k} + a_{k} = 4a_{k} \\
\sum_{a \in \mathbb{N}, 5 \leq a < 10} \left(2a + 1 \right) =& 11 + 13 + 15 + 17 + 19
\end{align}
$$
特に定数の和のときは
$$
\sum_{k=1}^{n} c = c \sum_{k=1}^{n} 1 = cn
$$

和の場合と同様に数列の積は次のように表す

$$
\underbrace{a_{1} \times a_{2} \times \cdots \times a_{n}}_{n\mbox{個}} = \prod_{i=1}^{n} a_{i}
$$

# 三角関数
三角関数は単位円上の$x$座標が$\cos$、$y$座標が$\sin$である。半径$r$の場合は

$$
x^{2} + y^{2} = r^{2} \\
\cos{\theta} = \frac{x}{r} \\
\sin{\theta} = \frac{y}{r} \\
$$
で定義される。

## 弧度法と度数法
三角関数の角度を表す際に弧度法と度数法が存在する。度数法とはいわゆる一般的な角度の定義のことである。他方で、弧度法とは、半径1の円の円周の長さで定義する。半径1の円周の長さは、

$$
2 \times 1 \times \pi = 2 \pi
$$
であるので、
$$
360^{\circ} = 2 \pi
$$

で定義される。弧度法の単位はラジアン(rad)。主要な角の度数法と弧度法の関係は以下の通り。

|      |            |                  |                  |                  |                  |                  |                  |              |                  |                  |                  |                  |                  |                  |                   |
|------|------------|------------------|------------------|------------------|------------------|------------------|------------------|--------------|------------------|------------------|------------------|------------------|------------------|------------------|-------------------|
|度数法 |$0^{\circ}$ |$30^{\circ}$      |$45^{\circ}$      |$60^{\circ}$      |$90^{\circ}$      |$120^{\circ}$     |$150^{\circ}$     |$180^{\circ}$ |$210^{\circ}$     |$225^{\circ}$     |$240^{\circ}$     |$270^{\circ}$     |$300^{\circ}$     |$315^{\circ}$     |$330^{\circ}$      |
|弧度法 |$0$         |$\frac{1}{6} \pi$ |$\frac{1}{4} \pi$ |$\frac{1}{3} \pi$ |$\frac{1}{2} \pi$ |$\frac{2}{3} \pi$ |$\frac{5}{6} \pi$ |$\pi$         |$\frac{7}{6} \pi$ |$\frac{5}{4} \pi$ |$\frac{4}{3} \pi$ |$\frac{3}{2} \pi$ |$\frac{5}{3} \pi$ |$\frac{7}{4} \pi$ |$\frac{11}{6} \pi$ |

角は一周回って$360^{\circ} = 2 \pi$であるので、定義上はいくらでも定義できるが、何周もすると扱いづらいので、

$$
 0 \leq \theta < 2\pi \\
 -\pi \leq \theta < \pi
$$
とすることが多い。このようにすると円周1周分に制限できる。角度は一周回って$2 \pi$であることから

```python
import math

# ラジアンから度数に変換する。
print('pi/6 = {} {} degree'.format(math.pi / 6, math.degrees(math.pi / 6)))
print('pi/4 = {} {} degree'.format(math.pi / 4, math.degrees(math.pi / 4)))
print('pi/3 = {} {} degree'.format(math.pi / 3, math.degrees(math.pi / 3)))

# 度数からラジアンに変換する
print('30 degree {} rad'.format(math.radians(30)))
print('45 degree {} rad'.format(math.radians(45)))
print('60 degree {} rad'.format(math.radians(60)))
```

# 指数・対数関数

## 指数とは
指数とは

$$
a^{x}
$$
のように表したときの$x$のことを指す。$a^{x}$とは

$$
a^{x} = \underbrace{a \times a \times \cdots \times a}_{x個}
$$
のことである。またxが正の整数としたときに、$a^{-x}$とは

$$
a^{-x} = \frac{1}{a^{x}}
$$
のことである。また、指数法則が成り立つ。

$$
\begin{align}
a^{x} a^{y} &= a^{x+y} \\
\left( a^{x} \right)^{y} &= a^{xy} \\
\frac{a^{x}}{a^{y}} &= a^{x-y} \\
\left( ab \right)^{x} &= a^{x} b^{x}
\end{align}
$$
この法則を元に、$a >0$、$x \in \mathbb{R}$としたとき、

$$
y = f \left( x \right) = a^{x}
$$
を指数関数という。
特に$x=\frac{1}{2}$のとき$\left(a^{\frac{1}{2}} \right)^{2} = a$であるから

$$
a^{\frac{1}{2}} = \sqrt{a}
$$
である。Pythonでは指数関数はアスタリスク2つで表す。例えば$2^{3}$は"2**3"と表す。

```python
import numpy as np
print(2**3)
print(2**(-1))
print(3**(1/2))
#np.sqrtは平方根
print(2**(np.sqrt(2)))
```

```python
# 図の表示用
import matplotlib.pyplot as plt
%matplotlib inline
x = np.linspace(-3, 3, 1000)
y = 2**x
fig, axe = plt.subplots()
axe.plot(x, y)
axe.grid()
```

```python
# 図の表示用
import matplotlib.pyplot as plt
%matplotlib inline
x = np.linspace(-3, 3, 1000)
y = 0.5**x
fig, axe = plt.subplots()
axe.plot(x, y)
axe.grid()
```

## 対数
指数関数

$$
y= a^{x}
$$

であり、$a$を$x$乗した値のことであった。今度は逆に$a$を何乗したら$x$になるのかという値を考えてみたい。そのような関数を対数関数と呼び

$$
y = \log_{a} x
$$
と書く。$a$を底、$x$を真数と呼ぶ。ただし$a>0$かつ$a \not=1$かつ$x > 0$とする。今仮に$a=2$とすると$x=8$のとき

$$
y = \log_{2} 8
$$
である。このとき2を3乗したら8であるから$y=3$である。$y=\log_{2} x$の関数は以下のような対応となる。

|    |              |              |              |    |    |    |    |         |
|----|--------------|--------------|--------------|----|----|----|----|---------|
|$x$ |$\frac{1}{8}$ |$\frac{1}{4}$ |$\frac{1}{2}$ |$1$ |$2$ |$4$ |$8$ |$\cdots$ |
|$y$ |$-3$          |$-2$          |$-1$          |$0$ |$1$ |$2$ |$3$ |$\cdots$ |

$y=\log_{a} x$と$y=a^{x}$は$y=x$について対称となる。  
また、対数には以下のような性質がある。

$$
\begin{align}
\log_{a}xy &= \log_{a}x + \log_{a}y \\
\log_{a}x^{y} &= y\log_{a}x \\
\end{align}
$$

特に以下のような性質がある。

$$
\begin{align}
\log{\prod_{i=1}^{n} a_{i}} &= \log{\left( a_{1} \times a_{2} \times \cdots \times a_{n} \right)} \\
                                           &= \log{a_{1}} + \log{a_{2}} + \cdots + \log{a_{n}} \\
                                           &= \sum_{i=1}^{n} \log{a_{i}}
\end{align}
$$

統計学などでこの性質はよく用いられる。  
また以下の公式を用いて底を変換できる(証明は省略)。

$$
\log_{a}b = \frac{\log_{c} b}{\log_{c} a}
$$

この対数関数は$x >0$で定義される。Pythonではmath関数でサポートされている。math(真数,底)で書く。numpyではサポートされていない。

```python
import math

print(math.log(8, 2))
print(math.log(0.25, 2))
print(math.log(7, 5))

# 図の表示用
import matplotlib.pyplot as plt
%matplotlib inline
x = np.linspace(0.125, 8, 1000)
y = [math.log(x_val, 2) for x_val in x.tolist()]
fig, axe = plt.subplots(figsize=(4,4))
axe.plot(x, y, label='log2 x')
x = np.linspace(-3, 3, 1000)
y = 2**x
axe.plot(x, y, label='2**x')
x = np.linspace(-3, 8, 1000)
axe.plot(x, x,linestyle='dashed', color='k')
axe.set_xlim([-3, 8])
axe.set_ylim([-3, 8])
axe.grid()
axe.legend()
```

# 微分法
## 平均変化率と微分法
$y= f\left( x \right)$について$x$から$x+h$に変化した時、下記の値を平均変化率という。

$$
\frac{f \left(x + h \right) - f \left(x \right)}{h} = \frac{yの変化量}{xの変化量}
$$
例えば、$y=\frac{1}{4} x^{2}$で$x=2$として$h=3, 1, 0.1, 0.001$としたときの平均変化率は次の通りとなる。

```python
import numpy as np
h_list = [3, 1,0.1, 0.001]
dev_list = []
for h in h_list:
    dev = (0.25 * (2+h)**2 - 0.25 * 2**2) / h
    print('h={}の平均変化率:{}'.format(h, dev))
    dev_list.append(dev)
    
# 図の表示用
import matplotlib.pyplot as plt
%matplotlib inline
x = np.linspace(0, 4, 1000)
y = 0.25 * x**2
fig, axe = plt.subplots()
axe.plot(x, y, label='y=1/4x**2', color='black')
for h, dev in zip(h_list, dev_list):
    y = dev * (x - 2) + 1
    axe.plot(x, y, label='h={}'.format(h))

axe.set_ylim(-1, 4)
axe.grid()
axe.legend()
```

```python
# 図の表示用
import matplotlib.pyplot as plt
%matplotlib inline
fig, axe = plt.subplots()
x = np.linspace(0, 4, 1000)
y = 0.25 * x**2
h = 0.001
axe.plot(x, y, label='0.25x**2', color='black')

dev = (0.25 * (2+h)**2 - 0.25 * 2**2) / h
y = dev * (x - 2) + 1
axe.plot(x, y, label='h={}'.format(h), color='red')

axe.set_ylim(-1, 4)
axe.grid()
axe.legend()
```

$h$の値が小さくなると徐々に傾きが$y=x^{2}$の接線の傾きに近づいていることがわかる。また、接線は、接点の付近では$y=x^{2}$の曲線とほぼ一致していることがわかる。すなわち$h$が小さくなったときの平均変化率は限りなく曲線の局所的な変化率に近づくことがわかる。このような局所的な変化率を求めることを微分と呼び、$y=f \left( x \right)$の微分のことを$\frac{dy}{dx}$、$f' \left( x \right)$、$\frac{d}{dx}f \left( x \right)$などと表す。特に$\frac{dy}{dx}=0$となり、符号変化がおこるところをプラスからマイナスに変わるところは、増加から減少に転じるところで、極大値、その逆を極小値という。

$$
\frac{dy}{dx} = \lim_{h \rightarrow 0} \frac{f \left(x + h \right) - f \left(x \right)}{h} 
$$

$x=a$のときの値

$$
\left. \frac{dy}{dx} \right|_{x=a} = f' \left( a \right)
$$
を$x=a$のときの微分係数と呼ぶ。

一般に$y= x^{a}$の微分は$y'=ax^{a-1}$であることはよく知られている。また微分の基本的な関係として以下の関係が成り立つ(定義から導かれる)。

$$
\begin{align} 
\left\{ f \left( x\right) + g \left( x\right) \right\}' &= f' \left( x\right) + g' \left( x\right) \\
\left\{k f \left( x\right)  \right\}' &= k f' \left( x\right) \\
\left\{ f \left( x\right)  g \left( x\right) \right\}' &= f' \left( x\right)  g \left( x\right) + f \left( x\right)  g' \left( x\right)
\end{align}
$$

## 合成関数の微分法

関数$y=f \left( t \right)$について、さらに$t$が$x$の関数であるとき、すなわち$t = g \left( x \right)$とすると以下の関係が成り立つ。

$$
\frac{dy}{dx} = \frac{dy}{dt} \frac{dt}{dx} = f' \left( t \right) g' \left( x \right)
$$

形式的に$\frac{dy}{dt}$と$\frac{dt}{dx}$を約分した形となっている。$y=t^{2}$、$t=x+1$としたとき

$$
\frac{dy}{dx} = 2t \times 1 = 2 x + 2
$$

これは$y$から$t$を消去した$y=\left( x+1 \right)^{2}$を微分した結果と一致していることがわかる。

## ネイピア数
ネイピア数とは$e=2.718281828459\cdots$という定数で$e^{x}$を微分したとき

$$
\left( e^{x} \right)' = e^{x}
$$
となることが知られている数である。またネイピア数を底にもつ対数$\log_{e}x$を自然対数と呼び$e$を省略して$\log x$、常用対数と区別して$\ln x$とかく。自然対数$y= \ln{x}$の微分は

$$
\frac{dy}{dx} = \frac{1}{x}
$$
となることが知られている。また底の変換公式を思い出すと

$$
\ln{x} = \frac{\log{x}}{\log{e}} \simeq 2.303 \log{x}
$$

```python
import numpy as np
print('ネイピア数:{}'.format(np.exp(1)))
print('log e:{}'.format(np.log10(np.exp(1))))
```

## 偏微分
これまでは一変数の場合の関数について考えたが$f \left( x_{1},x_{2},\cdots,x_{i},\cdots,x_{n} \right)$のような多変数関数を考える。このとき$x_{i}$以外の変数を全て定数とみて、$x_{i}$について微分することを偏微分と呼び$\frac{\partial f}{\partial x_{i}}$のように書く。$f_{x_{i}}$のように書く記法もあるが、機械学習ではあまり使われないので本講座では$\frac{\partial f}{\partial x_{i}}$で記載する。

$$
\frac{\partial f}{\partial x_{i}} = \lim_{h \rightarrow 0} \frac{f \left( x_{1},x_{2},\cdots,x_{i} + h,\cdots,x_{n} \right) - f \left( x_{1},x_{2},\cdots,x_{i},\cdots,x_{n} \right)}{h}
$$


### 例
以下のような関数を考える

$$ 
f\left(x_{1}, x_{2}, x_{3} \right) = x_{1}^{2} + 2 x_{2}^{3} + x_{1}x_{3}
$$

この関数をそれぞれの変数で偏微分すると

$$
\frac{\partial f}{\partial x_{1}} = 2 x_{1} + x_{3}, \ \ \frac{\partial f}{\partial x_{2}} = 6 x_{2}^2 , \ \ \frac{\partial f}{\partial x_{3}} = x_{1}
$$

$\left(x_{1}, x_{2}, x_{3} \right) = \left(1,1,1 \right)$の偏微分係数は

$$
\left. \frac{\partial f}{\partial x_{1}} \right|_{\left(x_{1}, x_{2}, x_{3} \right) = \left(1,1,1 \right)} = 3 \\
\left. \frac{\partial f}{\partial x_{2}} \right|_{\left(x_{1}, x_{2}, x_{3} \right) = \left(1,1,1 \right)} = 6 \\

## ベクトルの大きさ(長さ)
ベクトルの大きさは原点から各成分の距離に対応していて

ベクトルの長さは$\left| \textbf{a} \right|$、$\left\| \textbf{a} \right\|$、またまた単に$a$と表すなど様々な書き方が存在する。また、大学数学以降ではこれ以外の距離の定義も存在するので、この距離の定義をユークリッド距離と呼ぶこともある。

## ベクトルの定数倍
ベクトル$
\textbf{a} = \left(
\begin{array}{c}
a_{1} \\
a_{2} \\
\cdots \\
a_{n}
\end{array}
\right)
$
の定数倍を実数$k$を用いて
$$

$$
\textbf{0} = \left(
\begin{array}{c}
0 \\
0 \\
\cdots \\
0
\end{array}
\right)
$$

# 基底とベクトル空間
一旦、ここまでのベクトルの定義は忘れて、ベクトル空間について定義しておく。ある集合を$V$とし、$^{\forall}a, ^{\forall}b, ^{\forall}c \in V$に対して、以下の演算が成り立つときに、$V$はベクトル空間という。ただし$^{\forall}k, ^{\forall}l \in \mathbb{R}$とする。

1. 加法が閉じていること。$a+b \in V$
2. 加法の結合律が成り立つ。$a + \left( b + c \right) =\left( a + b \right) + c$
3. 加法の交換律が成り立つ。$a + b = b + a$
4. 加法の0元が存在すること。$^{\exists}\textbf{0} \in V, a + \textbf{0} = \textbf{0} + a = a$
5. 加法の逆元が存在すること。 $^{\exists}\textbf{0}, ^{\exists}\left( -a \right) \in V, a + \left( -a \right) = \left( -a \right) + a = \textbf{0}$
6. スカラー積が閉じていること。$ka \in V$
7. スカラーの結合律が成り立つ。$\left(kl \right)a =\left(k \right)la  $
8. スカラー積の単位元が存在すること。$1 \in \mathbb{R}, 1a=a $
9. スカラー積の0元が存在すること。 $0 \in \mathbb{R}, 0a=0$
10. スカラー積の分配則が成り立つ。 $k \left( a + b \right) = ka + kb$

これは、ここまでみてきたベクトルに限らず、この性質を満たす集合のことをベクトル空間と呼び、その元ををベクトルを呼ぶ。ここまでみてきたベクトルのことを対比として幾何ベクトルと呼ぶ。当然ここまでみてきた幾何ベクトルもベクトル空間の性質を満たしている。ここからは一旦幾何ベクトルのことについて扱うことにする。

## ベクトルの一次独立について
ベクトルが$\textbf{a}_{1},\textbf{a}_{2}, \cdots, \textbf{a}_{n} \in V$が一次独立であるとは、ある実数$c_{1}, c_{2}, \cdots , c_{n}$について

$$
c_{1} \textbf{a}_{1} + c_{2} \textbf{a}_{2} + \cdots + c_{n} \textbf{a}_{n} = \textbf{0}
$$
を満たす$c_{1}, c_{2}, \cdots , c_{n}$が

$$
c_{1} = c_{2} = \cdots = c_{n} = 0
$$
のみであることを言う。一次独立でないことを一次従属という。この定義に基づくと一次従属であれば

$$
- c_{n} \textbf{a}_{n} =  c_{1} \textbf{a}_{1} + c_{2} \textbf{a}_{2} + \cdots + c_{n-1} \textbf{a}_{n-1}
$$
を満たす$c_{1} = c_{2} = \cdots = c_{n} = 0$以外の$c_{1}, c_{2}, \cdots , c_{n}$の組み合わせが存在することになる。言い換えれば、ベクトルの一つが他のベクトルのペアで説明できるときに一次従属という。逆に説明できない時を一次独立という。$n$次元のベクトルについては、$n+1$個以上のベクトルは必ず一次従属となる。

特に二次元の場合は

$$
\textbf{a}_{2} = k \textbf{a}_{1}
$$
が$k=0$以外でも成り立つものが存在する。すなわち二次元の場合には一次従属であれば平行なベクトルということになる。

## 基底と次元
あるベクトル空間$\textbf{a}_{1}, \textbf{a}_{2}, \cdots, \textbf{a}_{n} \in V$とし、以下の条件を満たすときに$\textbf{a}_{1}, \textbf{a}_{2}, \cdots, \textbf{a}_{n}$は基底という。

1. $^{\forall}\textbf{x} \in V, ^{\exists}k_{1},^{\exists}k_{2}, \cdots, ^{\exists}k_{n} \in \mathbb{R}, \textbf{x} = k_{1}\textbf{a}_{1}+ k_{2}\textbf{a}_{2}+ \cdots + k_{n}\textbf{a}_{n}$
2. $\textbf{a}_{1}, \textbf{a}_{2}, \cdots, \textbf{a}_{n}$は一次独立。

この基底の数のことを次元と呼ぶ。この例では$n$個のベクトルが基底に準備されているので、$n$次元ベクトル空間と呼ぶ。特に実数の上で成り立つベクトル空間のことを実ベクトル空間(複素ベクトル空間とよばれるものもあるが、機械学習ではあまり用いられないのであまり気にしなくて良い)、実ベクトル空間全体のことを$\mathbb{R}^{n}$と表記する。  

### 基底の例
$\mathbb{R}^{n}$を考える。$n$個のベクトル

$$
\textbf{e}_{1} = \left(
\begin{array}{c}
1 \\
0 \\
\cdots \\
0
\end{array}
\right)
,
\textbf{e}_{2} = \left(
\begin{array}{c}
0 \\
1 \\
\cdots \\
0
\end{array}
\right)
,
\cdots
\textbf{e}_{n} = \left(
\begin{array}{c}
0 \\
0 \\
\cdots \\
1
\end{array}
\right)
$$
は基底となる。事実、$k_{1},k_{2}, \cdots, ,k_{n} \in \mathbb{R}$

$$
\textbf{x} = \sum_{i=1}^{n} k_{i}\textbf{e}_{i} = \left(
\begin{array}{c}
k_{1} \\
k_{2} \\
\cdots \\
k_{n}
\end{array}
\right)
$$
となり、明らかに$\left(
\begin{array}{c}
k_{1} \\
k_{2} \\
\cdots \\
k_{n}
\end{array}
\right)$
は$\mathbb{R}^{n}$全体を説明でき、かつ、これが$\textbf{0}$となるのは$k_{1}=k_{2}=\cdots=k_{n}=0$のみであるので、一次独立であることがわかる。

特に2次元の場合、

$$
\textbf{e}_{1} =\left(
\begin{array}{c}
1 \\
0 
\end{array}
\right), 
\textbf{e}_{2} =\left(
\begin{array}{c}
0 \\
1 
\end{array}
\right)
$$
となり、$\textbf{e}_{1}$は$x$軸方向の単位ベクトル、$\textbf{e}_{2}$は$y$軸方向の単位ベクトルとなり、$x$成分、$y$成分をさす。三次元の場合は同様に、$x,y,z$方向の成分の単位ベクトルのことを差す。

特に、基底の中でも基底のノルムが全て1、互いに直交する基底のことを正規直交基底という。この例も正規直交基底である。

# 確率とベイズの定理について
## 確率の性質
確率においては、起こりうることを事象と呼ぶ。例えば、サイコロを投げた場合の出た目の数の場合1、2などのことを事象という。事象のすべての集まり(集合)のことを標本空間と呼び

$$
\Omega=\left\{1,2,3,4,5,6 \right\}
$$
などと書く。
ここで事象$A$が起こる確率のことを以下のように定義する。

$$
P \left( A \right) = \frac{\mbox{事象}A\mbox{が起こるすべての場合の数}}{\mbox{起こりうるすべての場合の数}}
$$

事象$A$をサイコロで偶数の目が出るという事象とすると

$$
P \left( A \right) = \frac{3}{6} = \frac{1}{2}
$$

事象$B$を3以上の目が出る確率とすると

$$
P \left( B \right) = \frac{4}{6} = \frac{2}{3}
$$
となる。

ここで$A$と$B$が同時に起こる確率$P \left(A \cap B \right)$は

$$
P\left(A \cap B \right) = \frac{2}{6} = \frac{1}{3}
$$

ここで$A$または$B$を満たす確率$P \left(A \cup B \right)$は

$$
P\left(A \cup B \right) = \frac{5}{6} 
$$

また、一般的に下記の性質が成り立つ

$$
\begin{align}
P\left(A \cup B \right) &= P \left( A \right) + P \left( B \right) - P\left(A \cap B \right) \\
                        &= \frac{2}{3} + \frac{1}{3} - \frac{1}{6} \\
                        &= \frac{5}{6}
\end{align}
$$

## 独立と条件付き確率
一般に二つの事象$A$、$B$が同時に起こる確率が、それぞれの確率となるとき事象$A$、$B$は独立であるという。すなわち下記が成り立つときのことをさす。

$$
P\left(A \cap B \right) = P \left( A \right)P \left( B \right)
$$

次に条件付き確率について紹介する。条件付き確率とは事象$A$が観測されたのちに事象Bが観測される確率のことで、次のように表す。

$$
P \left( B| A \right) = \frac{P\left(A \cap B \right)}{P \left( A \right)}
$$

次のような例を考える。

ジョーカーのないトランプからハートのエースを引く事象を$B$とすると

$$
P \left( B \right) = \frac{1}{52}
$$

ここでいま事前にハートを引いていることがわかっていたとし、ハートを引くという事象を$A$とする。
ハートを引いたと分かっていれば

$$
P \left( B| A \right) = \frac{1}{13}
$$
となりそうである。これを定義に従って求めると$P\left( A \right)=\frac{1}{4}$であるから、

$$
P \left( B| A \right) = \frac{P\left(A \cap B \right)}{P \left( A \right)} = \frac{\frac{1}{52}}{\frac{1}{4}} = \frac{1}{13}
$$
と求まる。

ここで、$A$と$B$が独立であるとき

$$
P \left( B| A \right) = \frac{P\left(A \cap B \right)}{P \left( A \right)} = \frac{P \left( A \right)P \left( B \right)}{P \left( A \right)} = P \left( B \right)
$$
が成り立ち、$B$は$A$の影響を受けないことが、独立の意味であることがわかる。



## ベイズの定理
条件付き確率の定義から

$$
\begin{align}
P \left( B| A \right) &= \frac{P\left(A \cap B \right)}{P \left( A \right)} \\
P \left( A| B \right) &= \frac{P\left(A \cap B \right)}{P \left( B \right)} \\
P\left(A \cap B \right) &= P \left( A| B \right) P \left( B \right)
\end{align}
$$
となることから、

$$
P \left( B| A \right) = \frac{P \left( A| B \right) P \left( B \right)}{P \left( A \right)}
$$

ここで、全事象$\Omega$を$n$個に$B_{1}, B_{2}, \cdots , B_{n} $のように分割する(それぞれ重なり合わないものとする)。すなわち$\Omega=B_{1} \cup B_{2} \cup \cdots \cup B_{n}$

$$
\begin{align}
P \left( B_{i}| A \right) &= \frac{P\left(A \cap B_{i} \right)}{P \left( A \right)} \\
                          &= \frac{P \left( A| B_{i} \right) P \left( B_{i} \right)}{P \left( A \right)}
\end{align}
$$
ここで
$$
\begin{align}
P \left( A \right) = P \left( A \cap \Omega \right) &= P \left( A \cap \left(B_{1} \cup B_{2} \cup \cdots \cup B_{n} \right) \right) \\
                                                    &= P \left( \left( A \cap B_{1} \right) \cup \left(A \cap \left( B_{2} \cup B_{3} \cup \cdots \cup B_{n} \right) \right) \right) \\
                                                    &= P \left( A \cap B_{1} \right) + P \left( \left( A \cap \left( B_{2} \cup B_{3} \cup \cdots \cup B_{n} \right) \right)  \right) \\
                                                    &= P \left( A \cap B_{1} \right) + P \left( A \cap B_{2} \right) + P \left( \left( A \cap \left( B_{3} \cup B_{4} \cup \cdots \cup B_{n} \right) \right)  \right) \\
                                                    &= \sum_{k=1}^{n} P \left( A \cap B_{k} \right) \\
                                                    &= \sum_{k=1}^{n} P \left( A| B_{k} \right) P \left( B_{k} \right)
\end{align}
$$
となることから、以下のようになる。

$$
\begin{align}
P \left( B_{i}| A \right) &= \frac{P \left( A| B_{i} \right) P \left( B_{i} \right)}{\sum_{k=1}^{n} P \left( A| B_{k} \right) P \left( B_{k} \right)}
\end{align}
$$

# 標本
何らかの観測をして$n$個の観測結果

$$
x_{1}, x_{2}, \cdots, x_{n}
$$
があったとき、平均は

$$
\bar{x} = \frac{1}{n} \sum_{i=i}^{n} x_{i}
$$
で定義される。標本の散らばり具合を標本分散は以下で定義される。

$$
s^{2} = \frac{1}{n} \sum_{i=1}^{n} \left(x_{i}-\bar{x} \right)^{2}
$$
標本分散の平方根を標本標準偏差と呼ぶ。

$$
s = \sqrt{\frac{1}{n} \sum_{i=1}^{n} \left(x_{i}-\bar{x} \right)^{2}}
$$


$$
x_{1}, x_{2}, \cdots, x_{n}
$$

の各値に標準化という処理をして、平均0、標準偏差1となるように次のような変換をする。。

$$
u_{i} = \frac{x_{i} - \bar{x}}{s}
$$
事実このようにすると

$$
\bar{u} = \frac{1}{n}\sum_{i=1}^{n} u_{i} = \frac{1}{n}\sum_{i=1}^{n} \frac{x_{i} - \bar{x}}{s} = \frac{1}{s}\left(\bar{x} - \bar{x} \right) = 0
$$
となり、平均が0となり、標準偏差は

$$
s_{u} = \sqrt{\frac{1}{n} \sum_{i=1}^{n} \left( u_{i} - \bar{u} \right)^{2}} = \sqrt{\frac{1}{n} \sum_{i=1}^{n}  u_{i}^{2}} = \sqrt{\frac{1}{n} \frac{1}{s^{2}} \left( x - \bar{x} \right)^{2}} = \sqrt{\frac{1}{s^{2}} s^{2}} = 1
$$
となることから、このような変換をすれば、平均0、分散1となる。

次に$n$個の2つのデータ
$$
\left(x_{1}, y_{1}\right), \left(x_{2}, y_{2}\right), \cdots, \left(x_{n}, y_{n}\right)
$$
がどのくらい相関があるかを示す量を標本相関係数と呼び以下のように定義する。
$$
r= \frac{\frac{1}{n}\sum_{i=1}^{n} \left( x - \bar{x} \right) \left( y - \bar{y} \right) }{\sqrt{\frac{1}{n} \sum_{i=1}^{n} \left( x_{i} - \bar{x} \right)^2} \sqrt{\frac{1}{n} \sum_{i=1}^{n} \left( y_{i} - \bar{y} \right)^2}} = \frac{\sum_{i=1}^{n} \left( x - \bar{x} \right) \left( y - \bar{y} \right) }{\sqrt{\sum_{i=1}^{n} \left( x_{i} - \bar{x} \right)^2} \sqrt{\sum_{i=1}^{n} \left( y_{i} - \bar{y} \right)^2}}
$$
平均は、numpyのmean、標準偏差はstd、相関係数はcorrcoefで計算できる。
また相関係数は、相関係数行列

$$
\left(
\begin{array}{cc}
r_{xx}  & r_{xy}  \\
r_{xy} &  r_{yy} 
\end{array}
\right)
=
\left(
\begin{array}{cc}
1  & r_{xy}  \\
r_{xy} &  1 
\end{array}
\right)
$$
で出力される。

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
x_y = np.array([[20,50,30,85,60,100,60,50,40,75,60,30,20],\
                [30,30,10,100,20,95,60,80,50,85,20,20,80]])

x = x_y[0]
y = x_y[1]
print(x)
print(y)
print('平均: x:{}, y:{}'.format(np.mean(x), np.mean(y)))
print('標準偏差: x:{}, y:{}'.format(np.std(x), np.std(y)))
print('相関係数: {}'.format(np.corrcoef(x_y)))
```
